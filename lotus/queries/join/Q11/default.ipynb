```python
import pandas as pd
import lotus
from lotus.models import LM

df_players = pd.read_csv("../../datasets/player_evidence_mine.csv").head(5)[['Player Name']]
df_teams = pd.read_csv("../../datasets/team_evidence.csv")[['Team Name']]

lm = LM(model="ollama/llama3.3:70b")
lotus.settings.configure(lm=lm)
```

```python
instruction = "The player {Player Name:left} was playing in team {Team Name:right} in 2015."
df = df_players.sem_join(df_teams, instruction)
```
